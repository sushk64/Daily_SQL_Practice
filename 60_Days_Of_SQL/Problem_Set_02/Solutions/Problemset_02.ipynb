## Problem 02: Query a list of CITY names from STATION for cities that have an even ID number. Print the results in any order, but exclude duplicates from the answer.

```python
%%sql

SELECT DISTINCT(CITY)
FROM station
WHERE ID%2=0
ORDER BY CITY
LIMIT 20;
```

## Problem 03: Find the difference between the total number of CITY entries in the table and the number of distinct CITY entries in the table

```python
%%sql

SELECT COUNT(CITY) - COUNT(DISTINCT(CITY)) AS Difference
FROM station;
```

## Problem 04: Query the two cities in STATION with the shortest and longest CITY names, as well as their respective lengths (i.e.: number of characters in the name). If there is more than one smallest or largest city, choose the one that comes first when ordered alphabetically.

```python
%%sql

SELECT CITY, MIN(LENGTH(CITY)) AS Shortest_and_Longest
FROM station
UNION ALL
SELECT CITY, MAX(LENGTH(CITY)) Shortest_Longest
FROM station
ORDER BY Shortest_Longest
LIMIT 2;
```

## Problem 05: Query the list of CITY names starting with vowels (i.e., a, e, i, o, or u) from STATION. Your result cannot contain duplicates

```python
%%sql

SELECT DISTINCT(city)
FROM station
WHERE LOWER(CITY) LIKE 'a%' or LOWER(CITY) LIKE 'e%' or LOWER(CITY) LIKE 'i%' or LOWER(CITY) LIKE 'o%'or LOWER(CITY) LIKE 'u%';
```
